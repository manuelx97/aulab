**Obiettivo:** Sviluppare un prompt che sia adatto alla traduzione fluida tra italiano e inglese.

```python

# TODO Inizializza il client di Open AI
from dotenv import load_dotenv
from openai import OpenAI

load_dotenv('.env')
client = OpenAI()
```

input_text = "Felpa tecnica leggera e traspirante, ideale per correre nelle giornate fresche."

complention = client.chat.completions.create(
    model="gpt-4o-mini",
    messages=[
        {"role": "system", "content": "Sei un traduttore professionale. Traduci il testo dall'italiano all'inglese o dall'inglese all'italiano in base alla lingua di partenza. Restituisci solo la traduzione, senza spiegazioni."},
        {"role": "user", "content": input_text},
    ],
)

print(complention.choices[0].message.content)
