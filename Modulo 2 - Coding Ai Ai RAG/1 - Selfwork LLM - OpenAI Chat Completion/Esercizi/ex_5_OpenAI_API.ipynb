**Traccia:** Scrivi un prompt che generi domande di comprensione su un breve testo.

**Obiettivo:** Sviluppare un prompt per creare domande educative basate su un contenuto.

```python

# TODO Inizializza il client di Open AI
import os
from dotenv import load_dotenv
from openai import OpenAI

load_dotenv('.env')
client = OpenAI()

```

```python
input_text = "L'acqua è essenziale per la vita. Gli esseri umani devono bere regolarmente per mantenersi idratati, e un'alimentazione equilibrata aiuta il corpo a funzionare bene durante la giornata."


complention = client.chat.completions.create(
    model="gpt-4o-mini",
    messages=[
        {"role": "system", "content": "Genera domande di comprensione su un breve testo. Crea domande chiare, utili e pertinenti al contenuto, in italiano."},
        {"role": "user", "content": input_text},
    ],
)

print(complention.choices[0].message.content)
```
