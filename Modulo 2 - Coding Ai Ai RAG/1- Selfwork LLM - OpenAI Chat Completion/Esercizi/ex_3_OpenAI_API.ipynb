```python

# TODO Inizializza il client di Open AI
import os
from dotenv import load_dotenv
from openai import OpenAI

load_dotenv('.env')
client = OpenAI()

```

```python
input_text = "Un nuovo studio dell'Università di Roma mostra che l'uso regolare di brevi pause durante il lavoro aumenta concentrazione e produttività. I ricercatori hanno osservato benefici soprattutto negli studenti e nei lavoratori d'ufficio."

complention = client.chat.completions.create(
    model="gpt-4o-mini",
    messages=[
        {"role": "system", "content": "Riassumi l'articolo in un solo paragrafo in italiano, con un massimo di 255 caratteri. Mantieni solo le informazioni essenziali e non aggiungere titoli o elenchi."},
        {"role": "user", "content": input_text},
    ],
)

print(complention.choices[0].message.content)
```
