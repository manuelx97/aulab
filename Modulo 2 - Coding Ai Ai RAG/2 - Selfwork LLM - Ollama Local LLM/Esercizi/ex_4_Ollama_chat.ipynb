**Traccia:** Trasforma un set di dati strutturati (come un dizionario Python) in una descrizione testuale completa.

**Obiettivo:** Convertire dati strutturati in una descrizione leggibile.

```python
# TODO Inizializza il modello locale con Ollama
import ollama

MODEL = "llama3.2"
```

```python
data = {"name": "Giovanni", "age": 30, "city": "Roma", "profession": "Ingegnere"}
input_text = f"Descrivi in modo naturale questi dati: {data}"

response = ollama.chat(
    model=MODEL,
    messages=[
        {"role": "system", "content": 'Trasforma i dati strutturati in una descrizione testuale completa e naturale in italiano. Non usare elenchi puntati.'},
        {"role": "user", "content": input_text},
    ],
)

print(response["message"]["content"])
```
