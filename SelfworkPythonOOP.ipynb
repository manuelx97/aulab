```python
from abc import ABC, abstractmethod


class Person(ABC):
    def __init__(self, name, surname, age):
        self.name = name
        self.surname = surname
        self.age = age

    @abstractmethod
    def describe(self):
        pass


class Dipendente(Person):
    def __init__(self, name, surname, age, azienda, stipendio):
        super().__init__(name, surname, age)
        self.azienda = azienda
        self.stipendio = stipendio

    def describe(self):
        return f"{self.name} {self.surname}, {self.age} anni, dipendente di {self.azienda}."


class Studente(Person):
    def __init__(self, name, surname, age, corso_di_studio, matricola):
        super().__init__(name, surname, age)
        self.corso_di_studio = corso_di_studio
        self.matricola = matricola

    def describe(self):
        return f"{self.name} {self.surname}, {self.age} anni, studente di {self.corso_di_studio}."


class Docente(Dipendente):
    def __init__(self, name, surname, age, azienda, stipendio, materia, ore_lezione):
        super().__init__(name, surname, age, azienda, stipendio)
        self.materia = materia
        self.ore_lezione = ore_lezione

    def describe(self):
        return f"{self.name} {self.surname} insegna {self.materia} e lavora per {self.azienda}."


class Freelance(Dipendente):
    def __init__(self, name, surname, age, azienda, stipendio, ambito, partita_iva):
        super().__init__(name, surname, age, azienda, stipendio)
        self.ambito = ambito
        self.partita_iva = partita_iva

    def describe(self):
        return f"{self.name} {self.surname} lavora come freelance nell'ambito {self.ambito}."


persone = [
    Dipendente("Marco", "Rossi", 34, "Azienda X", 2500),
    Studente("Anna", "Bianchi", 21, "Informatica", "S12345"),
    Docente("Luca", "Verdi", 45, "Istituto Y", 3200, "Python", 18),
    Freelance("Sara", "Neri", 29, "Studio Z", 2800, "Web Design", "IT123456789")
]

for persona in persone:
    print(persona.describe())
```
